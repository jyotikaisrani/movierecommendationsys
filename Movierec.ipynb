```python
moviemeta = pd.read_csv('/content/drive/MyDrive/dataset/movies_metadata.csv' , low_memory=False)
moviemeta.head(4)
```

```python

moviemeta.info()
```

```python
meanvote = moviemeta['vote_average'].mean()
print(meanvote)
```

```python
minimumvote = moviemeta['vote_count'].quantile(0.50)
print(minimumvote)
```

```python
q_movies =moviemeta.copy().loc[moviemeta['vote_count']>=minimumvote]
q_movies.shape
```

```python
def weighed_rating(x, minimumvote=minimumvote, meanvote= meanvote):
  voters=x['vote_count']
  avg_vote=x['vote_average']
  return (voters/(voters+minimumvote)*avg_vote) + (minimumvote/(minimumvote+voters)*meanvote)
```

```python
q_movies['score'] = q_movies.apply(weighed_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score' , ascending=False)
```
